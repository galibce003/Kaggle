## For categorical features

```python
categorical_features = [i for i in df.columns if df[i].dtype == 'O']
categorical_features
```

<br/>

#### List of categorical features having NaN values

```python
categorical_features_nan = [i for i in categorical_features if df[i].isnull().sum()>0]
categorical_features_nan
```

<br/>

#### Replace the NaN values with **'Missing'**

```python
df[categorical_features_nan] = df[categorical_features_nan].fillna('Missing')
```

```python
df[categorical_features_nan].isnull().sum()
```

<br/>

## List of numerical features having NaN values

```python
numerical_features_nan = [i for i in df.columns if df[i].dtype != 'O' and df[i].isnull().sum()]
numerical_features_nan
```

<br/>

#### replace the NaN values with median

```python
for i in numerical_features_nan:
    median = df[i].median()
    
    df[i].fillna(median, inplace = True)
```

```python
df[numerical_features_nan].isnull().sum()
```

<br/>

## List of the temporal features

```python
temporal_features = [i for i in df.columns if 'Yr' in i or 'Year' in i]
temporal_features
```

```python
for i in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[i] = df['YrSold'] - df[i]
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

<br/>

## Converting into log-normal distribution from skewed distribution

```python
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for i in num_feature:
    df[i] = np.log(df[i])
```

<br/>

## Rare categorical features
If any unique value of a feature presents **less 1%**  of the total feature, we will consider it as a rare categorical feature.

```python
categorical_features
```

<br/>

#### Replace the rare categories with **'Rare'**

```python
for feature in categorical_features:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp > 0.01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], 'Rare')
```

```python
temp_df
```

```python
df['MSZoning'].unique()
```

<br/>

## Labeling the categorical categorical_features

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head()
```

<br/>

## Scaling

```python
feature_scale = [feature for feature in df.columns if feature not in ['Id', 'SalePrice']]

scaler = MinMaxScaler()
scaler.fit(df[feature_scale])
```

<br/>

## Return it into Dataframe again
**Note:**  
Normally when wi apply scaling, the dataset gets converted into an array. So all we did here just convert the array into dataframe again with concatinating the **Id** and **SalePrice** feature.

```python
scaled_df = pd.concat([df[['Id', 'SalePrice']].reset_index(drop = True),
                      pd.DataFrame(scaler.transform(df[feature_scale]),
                                  columns = feature_scale)], axis = 1)
```

```python
scaled_df.head()
```

<br/>

## Saving the scaled dataset 

```python
scaled_df.to_csv('X_train_scaled.csv', index = False)
```
