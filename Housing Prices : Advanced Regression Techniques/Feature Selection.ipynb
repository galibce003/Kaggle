## Importing necessary modules

```python
import pandas as pd                    #Data Manipulation
import numpy as np                     #Linear Algebra
import matplotlib.pyplot as plt        #Visualization                             
import seaborn as sns                  #Visualization
%matplotlib inline 

from sklearn.linear_model import Lasso
from sklearn.feature_selection import RFE
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

pd.pandas.set_option('display.max_columns', None)         #Visualize all the columns
```

<br/>

## Load the dataset
This time I am going to use the dataset processed from Feature Engineering

```python
df = pd.read_csv('X_train_scaled.csv')
df.head()
```

<br/>

## Splitting into Explanatory and Response variable

```python
X_train = df.drop(['Id', 'SalePrice'], axis = 1)
y_train = df[['SalePrice']]
```

<br/>

## Feature Selection by Lasso

```python
feature_sel_model = SelectFromModel(Lasso(alpha = 0.0003, random_state = 0))
feature_sel_model.fit(X_train, y_train)
```

<br/>

### Boolean Mask of the selected feature

```python
feature_sel_model.get_support()
```

<br/>

### Features which got selected

```python
selected_feature = X_train.columns[(feature_sel_model.get_support())]
selected_feature
```

<br/>

### Length of the selected features

```python
print("Total features : {}".format(X_train.shape[1]))
print("Selected features : {}".format(len(selected_feature)))
```

<br/>

### Storing the selected features in X_train_lasso

```python
X_train_lasso = X_train[selected_feature]
X_train_lasso.head()
```

<br/>

### Model Build UP

```python
lr1 = LinearRegression()
model1 = lr1.fit(X_train_lasso, y_train)

# Prediction
y1_pred = lr1.predict(X_train_lasso)

# Accuracy Score
r2_score(y_train, y1_pred)*100
```

<br/>

## Feature Selection with RFE

```python
rfe = RFE(estimator = LinearRegression(), n_features_to_select = 25, verbose = 1)
rfe.fit(X_train, y_train)
```

<br/>

### Prediction and Accuracy Score

```python
y2_pred = rfe.predict(X_train)

r2_score(y_train, y2_pred)*100
```

<br/>

### List of the selected features

```python
X_train_ref = X_train.columns[rfe.support_]
X_train_ref
```

<br/>

### Diving into a bit deeper with the features along with RFE

```python
print(dict(zip(X_train.columns, rfe.ranking_)))
```
